# Método del Bisección

```python
def funcion(h):
    return 10*(0.5*(math.pi)*pow(1,2) - pow(1,2)*math.asin(h/1) - h*pow((pow(1,2) - pow(h,2)),1/2)) - 12.4

resultados =[['n','a','b','p','f(p)','error']]

a=0
b=1
n=50
error=0.01
print("f(",a,") = ", funcion(a))
print("f(",b,") = ", funcion(b))
for i in range(1,50):
        p = (a+b)/2
        f=funcion(p)
        errorCalc = abs((a-b)/2)
        
        
        resultados.append([i,a,b,p,f,errorCalc])
        
        if(funcion(a)*funcion(p)<0):
            b=p
        else:
            a=p
            
        
        if(abs(errorCalc) < error):
            break
for fila in resultados:
    dato1, dato2, dato3, dato4, dato5, dato6 = fila
    print ("{:<2} {:<25} {:<25} {:<25} {:<25} {:<25}".format(dato1, dato2, dato3, dato4, dato5, dato6))
```

# Método Punto Fijo

```python
# Ingresar la función
def funcion(x):
    return (300 + 201.0625*(1-pow(math.e,-0.4*x)))/(80.425)

resultados =[['n','p','f(p)','|p-po|']]

# Colocar en a y b los intervalos, si fuera el caso
a = 5
b = 6

# Colocar el valor de p inicial
p = a

# Colocar la cantidad de iteraciones
n = 250

# Colocar la tolerancia
error = 0.01

print("f(",a,") = ", funcion(a))
print("f(",b,") = ", funcion(b))

for i in range(0,n):
    f=funcion(p)
    errorCalc = abs(p-f)

    resultados.append([i,p,f,errorCalc])

    if(errorCalc > error):
        p=f
        if(p==0):
            print("Indefinido")
    else:
        break

for fila in resultados:
    dato1, dato2, dato3, dato4 = fila
    print ("{:<2} {:<25} {:<25} {:<25}".format(dato1, dato2, dato3, dato4))
```

### Calculo de Derivada

```python
# Ingresar la función
def funcion(x):
    return pow(x,2) - 6

# Ingrese la derivada
def derivada(x):
    return 2*x

# Muestra iteracion - Valor P - Valor Funcion - Valor Derivada - Valor Tolerancia
resultados =[['n','p','f(p)','f`(p)','P','|p-po|']]

# Colocar el valor de p inicial
p = 1

# Colocar la cantidad de iteraciones
n = 250

# Colocar la tolerancia
error = 0.01

for i in range(0,n):
    f=funcion(p)
    d=derivada(p)
    h = p - (f/d)
    errorCalc = abs(h - p)

    resultados.append([i,p,f,d,h,errorCalc])

    if(errorCalc > error):
        p=h
        if(p==0):
            print("Indefinido")
    else:
        break

for fila in resultados:
    dato1, dato2, dato3, dato4, dato5, dato6 = fila
    print ("{:<2} {:<25} {:<25} {:<25} {:<25} {:<25}".format(dato1, dato2, dato3, dato4, dato5, dato6))
```
